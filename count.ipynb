```python
import os

def count_total_lines(folder_path):
    total_lines = 0
    for root, _, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    lines = f.readlines()
                    total_lines += len(lines)
            except UnicodeDecodeError:
                print(f"‚ö†Ô∏è B·ªè qua file kh√¥ng ph·∫£i text ho·∫∑c b·ªã l·ªói m√£ h√≥a: {file_path}")
            except Exception as e:
                print(f"‚ùå L·ªói khi ƒë·ªçc {file_path}: {e}")
    return total_lines

# S·ª≠ d·ª•ng:
folder = "data/processed-local"
total = count_total_lines(folder)
print(f"T·ªïng s·ªë d√≤ng trong t·∫•t c·∫£ c√°c file ·ªü '{folder}': {total}")
```

```python
import os
import csv
from collections import defaultdict

def group_files_by_column_count(folder_path):
    col_groups = defaultdict(list)

    for root, _, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    reader = csv.reader(f)
                    for row in reader:
                        num_cols = len(row)
                        if num_cols > 0:
                            col_groups[num_cols].append(file_path)
                        break  # ch·ªâ ƒë·ªçc d√≤ng ƒë·∫ßu ti√™n
            except Exception as e:
                print(f"‚ö†Ô∏è B·ªè qua file l·ªói: {file_path} ({e})")

    return col_groups

# S·ª≠ d·ª•ng
folder = "data/raw"
column_groups = group_files_by_column_count(folder)

# In k·∫øt qu·∫£
if column_groups:
    print("üìä Ph√¢n nh√≥m file theo s·ªë c·ªôt:\n")
    for col_count in sorted(column_groups):
        print(f"S·ªë c·ªôt: {col_count} ‚Äî {len(column_groups[col_count])} file(s):")
        for file in column_groups[col_count]:
            print(f"  - {file}")
        print()
else:
    print("‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y file h·ª£p l·ªá n√†o trong th∆∞ m·ª•c.")
```

```python
import os
import csv

# Danh s√°ch file c·∫ßn ƒë·∫øm d√≤ng
files_to_count = [
    # C√°c file c√≥ 56 c·ªôt
    "data/raw/travistorrent_28_8_2015.csv",
    "data/raw/travistorrent_5_3_2016.csv",

    # File c√≥ 66 c·ªôt
    "data/raw/final-2017-01-25.csv",
]

# ƒê·∫øm d√≤ng m·ªôt file
def count_lines(file_path):
    count = 0
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            for _ in f:
                count += 1
        return count - 1  # Tr·ª´ d√≤ng header
    except Exception as e:
        print(f"‚ö†Ô∏è L·ªói khi ƒë·ªçc file: {file_path} ({e})")
        return 0

# ƒê·∫øm t·ªïng d√≤ng c·ªßa t·∫•t c·∫£ file c√≥ s·ªë c·ªôt = 62
def count_lines_for_column_count(folder_path, target_col_count):
    total = 0
    for root, _, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    reader = csv.reader(f)
                    first_row = next(reader)
                    if len(first_row) == target_col_count:
                        count = sum(1 for _ in reader)
                        total += count
            except Exception as e:
                print(f"‚ö†Ô∏è B·ªè qua file l·ªói: {file_path} ({e})")
    return total

# ƒê·∫øm d√≤ng cho c√°c file ri√™ng l·∫ª
print("üìÅ S·ªë d√≤ng c·ªßa c√°c file c·ª• th·ªÉ:")
for file in files_to_count:
    print(f"{file}: {count_lines(file)} d√≤ng")

# ƒê·∫øm d√≤ng cho t·∫•t c·∫£ file c√≥ 62 c·ªôt
total_62_cols = count_lines_for_column_count("data/raw", 62)
print(f"\nüìä T·ªïng s·ªë d√≤ng (tr·ª´ header) c·ªßa t·∫•t c·∫£ file c√≥ 62 c·ªôt: {total_62_cols} d√≤ng")
```
