```python
import os
import pandas as pd

# ƒê∆∞·ªùng d·∫´n t·ªõi th∆∞ m·ª•c ch·ª©a file
folder_path = 'data/processed/main'

# T·∫°o dataframe th·ªëng k√™
summary = []
total_rows_all_files = 0  # bi·∫øn t·ªïng s·ªë d√≤ng

# Duy·ªát qua t·∫•t c·∫£ c√°c file trong th∆∞ m·ª•c
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        try:
            df = pd.read_csv(file_path)
            total_rows = len(df)
            total_rows_all_files += total_rows  # c·ªông d·ªìn s·ªë d√≤ng
            if 'build_failed' in df.columns:
                value_counts = df['build_failed'].value_counts(normalize=True).to_dict()
                percent_0 = round(value_counts.get(0, 0) * 100, 2)
                percent_1 = round(value_counts.get(1, 0) * 100, 2)
            else:
                percent_0 = percent_1 = None
        except Exception as e:
            total_rows = None
            percent_0 = percent_1 = None
            print(f"‚ö†Ô∏è L·ªói khi ƒë·ªçc {file_name}: {e}")

        summary.append({
            'File': file_name,
            'S·ªë d√≤ng': total_rows,
            '% build_failed = 0': percent_0,
            '% build_failed = 1': percent_1
        })

# Hi·ªÉn th·ªã b·∫£ng th·ªëng k√™
summary_df = pd.DataFrame(summary)
summary_df = summary_df.sort_values(by='S·ªë d√≤ng', ascending=False).reset_index(drop=True)
display(summary_df)

# Hi·ªÉn th·ªã t·ªïng s·ªë d√≤ng
print(f"üìäTotal rows all files: {total_rows_all_files:,}")
```
